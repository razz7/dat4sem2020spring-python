```python
# Exercise part 2
import kkdata
data = kkdata.STATISTICS
years = data.keys()
population_by_year = {}
for year in years:
    areas = data[year].keys()
    for area in areas:
        ages = data[year][area].keys()
        for age in ages:
            country_codes = data[year][area][age].keys()
            for country_code in country_codes:
                population_by_year.setdefault(year,0)
                population_by_year[year] += data[year][area][age][country_code]
plt.figure()
plt.plot(list(population_by_year.keys()),list(population_by_year.values()))
                
```

# Location data with geoJson

```python
# Getting the data on ghettos
from mylib import geodata_util as geo

url = 'http://wfs-kbhkort.kk.dk/k101/ows?service=WFS&version=1.0.0&request=GetFeature&typeName=k101:f_udsatte_byomraader&outputFormat=application%2Fjson&SRSNAME=EPSG:4326'
lat, long = geo.get_city_location('Copenhagen')
geojson = geo.get_geojson(url)
```

```python
import folium
import urllib
cph_map = folium.Map(location=(lat,long), zoom_start=12)
folium.GeoJson(data=geojson).add_to(cph_map)
cph_map.save('ghetto.html')
cph_map
```
